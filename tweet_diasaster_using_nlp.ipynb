```python
import pandas as pd
import numpy as np
import re
import nltk
```

```python
data = pd.read_csv('train.csv')
```

```python

data
```

```python
data = data.drop(['keyword','id','location'],axis=1)
```

```python
data
```

```python
data['text']
```

```python
data['text'] = data['text'].apply(lambda x: x.lower())
```

```python
data['text']
```

```python
data['text'] = data['text'].apply(lambda x:re.sub(r"http\S+","",x))
```

```python
data['text'] = data['text'].apply(lambda x:re.sub(r'[^a-zA-Z]',' ',x))
```

```python
data['text']
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt')
```

```python
stop_words = stopwords.words('english')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    words = nltk.word_tokenize(text)
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    return ' '.join(words)
```

```python
data['cleaned_text'] = data['text'].apply(preprocess_text)
```

```python
data
```

```python
data=data.drop(['text'],axis=1)
```

```python
data
```

```python
from gensim.models import Word2Vec
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data['cleaned_text'], data['target'], test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
idf_pip=Pipeline(
                [
                    ('tf_idf',TfidfVectorizer(ngram_range=(1,1))),
                    ('model',LogisticRegression(C=.8,solver='sag',max_iter=1000))
                ]
                )
idf_pip.fit(X_train,y_train)
```

```python
print("TfidfVectorizer Model Train Score is :",idf_pip.score(X_train,y_train))
print("TfidfVectorizer Model Test Score is :",idf_pip.score(X_test,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(ngram_range=(1,1))),
    ('model', RandomForestClassifier())
])

rf_pipeline.fit(X_train, y_train)

print("Random Forest Model Train Score is :", rf_pipeline.score(X_train, y_train))
print("Random Forest Model Test Score is :", rf_pipeline.score(X_test, y_test))
```

```python

from sklearn.model_selection import GridSearchCV

# Define the parameter grid for Random Forest
param_grid = {
    'model__n_estimators': [50, 100, 200],
    'model__max_depth': [None, 10, 20],
    'model__min_samples_split': [2, 5, 10]
}

# Create GridSearchCV object
grid_search = GridSearchCV(rf_pipeline, param_grid, cv=5, scoring='accuracy')

# Fit the grid search to the training data
grid_search.fit(X_train, y_train)

# Print the best parameters and best score
print("Best Parameters: ", grid_search.best_params_)
print("Best Score: ", grid_search.best_score_)

# Evaluate the best model on the test set
best_model = grid_search.best_estimator_
print("Test Accuracy with Best Model: ", best_model.score(X_test, y_test))
```

```python
from sklearn.metrics import classification_report
# Predict on the test set using the best model
y_pred = best_model.predict(X_test)

# Generate and print the classification report
print(classification_report(y_test, y_pred))
```

```python
# Assuming 'y_pred' contains the predicted target values
predictions_df = pd.DataFrame({
    'id': range(len(y_pred)),  # Creating a simple range of IDs
    'target': y_pred
})

# Save the DataFrame to a CSV file
predictions_df.to_csv('predictions.csv', index=False)
```
